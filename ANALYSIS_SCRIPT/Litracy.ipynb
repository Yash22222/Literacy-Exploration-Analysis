## 3. PATTERN ANALYSIS

```python
# CHecking Rates Feom Selected States

selected_states = ['Kerala', 'Tamil Nadu', 'Karnataka', 'Andhra Pradesh', 'Maharashtra']
```

```python
plt.figure(figsize=(10, 6))
for state in selected_states:
    state_data = df[df['Country/ States/ Union Territories Name'] == state]
    plt.plot(state_data.columns[2:], state_data.iloc[0, 2:], label=state, marker='o')
plt.xlabel('Years')
plt.ylabel('Literacy Rate (Persons)')
plt.title('Literacy Rates Trend Over Years With Selected States')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
plt.bar(df['Country/ States/ Union Territories Name'], df['Literacy Rate (Persons) - Total - 2011'], color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('States/Union Territories')
plt.ylabel('Literacy Rate (Persons) - Total - 2011')
plt.title('Total Literacy Rates in 2011 Across Different States & Union Territories')
plt.tight_layout()
plt.show()
```

```python
# Now For ONe Selected State

selected_states = ['Maharashtra']
plt.figure(figsize=(10, 6))
for state in selected_states:
    state_data = df[df['Country/ States/ Union Territories Name'] == state]
    plt.plot(state_data.columns[2:-1], state_data.iloc[0, 2:-1], label=state, marker='o')
plt.xlabel('Years')
plt.ylabel('Literacy Rate (Persons)')
plt.title('Literacy Rates Trend Over Years With Selected States')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
